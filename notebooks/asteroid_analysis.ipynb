```python
import pandas as pd 
import seaborn as sns
import matplotlib as plt

# Read the data
df=pd.read_csv("../data/processed/asteroids_final.csv")
df.head()
```

# ANALYZE THE DATA

```python
# How many asteroids were examined in total?
print(f"Number of astreoids: {len(df)}")
```

```python
# How many of them are potentially hazardous?
hazardous_count= len(df[df["hazardous_flag"]==1])
print(f"Number of potentially hazardous asteroids: {hazardous_count}")
```

```python
# What is the average approach distance in km?
print(f"Average distance: {df["miss_distance_km"].mean()} km")
```

```python
# What is the average velocity in km/h?
print(f"Average velocity: {df["relative_velocity_km_s"].mean()} km/s")
```

```python
# Which is the biggest asteroid?
biggest=df.loc[df["average_diameter"].idxmax()]
print(f"Biggest asteroid: {biggest["name"]}")
print(f"Average diameter: {biggest["average_diameter"]} km")
```

```python
# Which asteroid passed closest to Earth?
closest=df.loc[df["miss_distance_km"].idxmin()]
print(f"Closest the Earth: {closest["name"]}")
print(f"Distance: {closest["miss_distance_km"]} km")
```

```python
# Which asteroid has the highest risk score?
higgest=df.loc[df["risk_score"].idxmax()]
print(f"Riskiest asteroid: {higgest["name"]}")
print(f"Risk score: {higgest["risk_score"]}")
print(f"Risk level: {higgest["risk_level"]}")
```
