```python
import pandas as pd 
import seaborn as sns
import matplotlib as plt

# Read the data
df=pd.read_csv("../data/processed/asteroids_final.csv")
df.head()
```

# ANALYZE THE DATA

```python
# How many asteroids were examined in total?
print(f"Number of astreoids: {len(df)}")
```

```python
# How many of them are potentially hazardous?
hazardous_count= len(df[df["hazardous_flag"]==1])
print(f"Number of potentially hazardous asteroids: {hazardous_count}")
```

```python
# What is the average approach distance in km?
print(f"Average distance: {df["miss_distance_km"].mean()} km")
```

```python
# What is the average velocity in km/h?
print(f"Average velocity: {df["relative_velocity_km_s"].mean()} km/s")
```

```python
# Which is the biggest asteroid?
biggest=df.loc[df["average_diameter"].idxmax()]
print(f"Biggest asteroid: {biggest["name"]}")
print(f"Average diameter: {biggest["average_diameter"]} km")
```

```python
# Which asteroid passed closest to Earth?
closest=df.loc[df["miss_distance_km"].idxmin()]
print(f"Closest the Earth: {closest["name"]}")
print(f"Distance: {closest["miss_distance_km"]} km")
```

```python
# Which asteroid has the highest risk score?
higgest=df.loc[df["risk_score"].idxmax()]
print(f"Riskiest asteroid: {higgest["name"]}")
print(f"Risk score: {higgest["risk_score"]}")
print(f"Risk level: {higgest["risk_level"]}")
```

```python
# Which month has more asteroid approaches?

import calendar

month_map = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

df['month_name'] = df['month'].map(month_map)

monthly_counts = df['month_name'].value_counts()
print("Asteroid Counts per Month:")
print(monthly_counts)
print("-----------------")

peak_month_num = df['month'].value_counts().idxmax()
peak_count = df['month'].value_counts().max()
peak_month_name = month_map[peak_month_num]

print(f"\nAnalysis Result:")
print(f"The highest asteroid traffic was observed in {peak_month_name} with {peak_count} total approaches.")
```

```python
# On which days are risky approaches concentrated?

risky_days = df[df['risk_level'].isin(['High', 'Critical'])]['day_name'].value_counts()
print("\nDays with most risky approaches:")
print(risky_days)
```

```python
# What is the distribution of asteroids by risk level?

risk_counts = df['risk_level'].value_counts()
print("Distribution of Asteroids by Risk Level:")
print(risk_counts)

risk_percentages = df['risk_level'].value_counts(normalize=True) * 100
print("\nPercentage Distribution (%):")
print(risk_percentages.round(2))
```

```python
# Is the average velocity of hazardous asteroids higher?
ave_vel_hazar=df.groupby("hazardous_flag")["relative_velocity_km_s"].mean()
print("\nAverage Velocity by Hazard Flag (0: Safe, 1: Hazardous):")
print(ave_vel_hazar)
```
